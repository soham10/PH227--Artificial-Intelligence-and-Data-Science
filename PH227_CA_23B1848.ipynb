```python
import numpy as np
import matplotlib.pyplot as plt
import time

np.random.seed(int(time.time())) #seeding
```

## Question 1 

```python
n = 10   #number of trials per sample
p = 0.5  #probability of success
size = 10000 #number of samples

#Bernaulli Trial Distribution
X = np.zeros(size)
for i in range(size):
    x = np.random.choice([1,0],n,p=[p,1-p])
    X[i] = np.sum(x)

#Binomial Distribution
Y = np.random.binomial(n,p,size)

#Plotting the Histogram
plt.hist(X,bins=100,alpha=0.5,density=True,color='blue',label='bernoulli')
plt.hist(Y,bins=100,alpha =0.6,density=True,color='red',label='binomial')
plt.xlabel('Success')
plt.ylabel('Frequency')
plt.title('Comparison of Binomial Distributions')
plt.legend()
plt.show()
```

## Question 2

```python
par_lambda = 0.9 # Lambda parameter
n = 10000 # Number of data points for discretization

x = np.linspace(0,10,n)  # Distribution of data points uniformly

p = par_lambda*np.exp(-1*par_lambda*x)  # probability for exponential distribution
p_norm = p/np.sum(p)      #normalization

X = np.random.choice(x,n,p=p_norm)

#Plot/histogram for exponential distribution
plt.hist(X,bins=100,alpha=0.5,density=True,color='blue')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Exponential Distribution')
plt.show()

cdf = np.cumsum(p_norm)  # Using in-built np.cumsum()

#Plot of CDF
plt.plot(x, cdf, label='CDF')
plt.xlabel('x')
plt.ylabel('Cumulative Value')
plt.title('Cumulative Distribution Function')
plt.legend()
plt.show()
```

## Question 3

```python
N = 10000 #Maximum number of dimensions to be plotted
n = np.zeros(N) #Initialising dimension array where n varies from 0 to N
angle = np.zeros(N) #Intialising angle array for corresponding n

# Calculating angle using dot product
for i in range(1,N):
    vec_1 = np.random.normal(0,1/np.sqrt(i),i)
    vec_2 = np.random.normal(0,1/np.sqrt(i),i)
    dot = np.dot(vec_1,vec_2)
    angle[i]  = np.arccos(dot/(np.linalg.norm(vec_1)*np.linalg.norm(vec_2)))
    n[i] = i

print(angle[N-1]*180/np.pi) # Angle at n = N

# Plot to demonstrate Law of Large Numbers
plt.plot(n,angle*180/np.pi) 
plt.xlabel('Dimensions')
plt.ylabel('Angle (degrees)')
plt.title('Law of Large Numbers')
plt.show()

print('As quite visible in the plot, the value of angle saturates to 90 degrees as N tends to infinity.')
```
